```python
import spacy
spacy.cli.download("en_core_web_sm")
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
text = "Apple is looking at buying a (U.K.) startup for $1 billion."
print(text)
```

```python
text = "Apple is looking at buying a (U.K.) startup for $1 billion.And i love U.K."
doc = nlp(text)
print(doc)
```

```python
print(len(text))
print(len(doc))
```

```python
for token in text:
  print(token)
```

```python
for token in doc:
    print(token.text)
```

```python
for token in text.split():
   print(token)
```

**Sentence Boundary Detection**

In NLP, sentence boundary detection, or SBD, is the identification of sentences in a text.

```python
for sent in doc.sents:
    print (sent)
```

```python
sentence1 = list(doc.sents)[0]
print (sentence1)
```

```python
token2 = sentence1[2]
print (token2)
```

**Part of Speech Tagging (POS)**

In the field of computational linguistics, understanding parts-of-speech is essential. SpaCy offers an easy way to parse a text and identify its parts of speech. Below, we will iterate across each token (word or punctuation) in the text and identify its part of speech.

```python
for token in sentence1:
    print (token.text, token.pos_, token.dep_)
```

```python
from spacy import displacy
displacy.render(sentence1, style="dep")
```
